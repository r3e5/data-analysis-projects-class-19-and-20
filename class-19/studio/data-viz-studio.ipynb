```python
# Here is the setup of the dataframe. Feel free to use .info() or .head() to 
## get a better understanding of what is inside the dataframe!

import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style

df = pd.read_csv("books.csv", on_bad_lines='skip')

years = []

for y in df['publication_date']:
    date = y.split("/")
    years.append(date[2])

df["year"] = years

df.columns
```

With your dataframe set up, time to start creating visualizations!

```python
# Use your Python skills to create a visualization here.
top_10_publishers = df.groupby('publisher')['title'].count().sort_values(ascending=False).head(10)

publishers = top_10_publishers.index
num_books = top_10_publishers.values

plt.bar(publishers, num_books)
plt.xticks(rotation=45, ha='right')

plt.xlabel('Publishers')
plt.ylabel('Number of Books')
plt.title('Number of Books Published by Top 10 Publishers')

plt.show()
```

```python
# Use your Python skills to create a visualization here.
top_10_publishers = df.groupby('publisher')['title'].count().sort_values(ascending=False).head(10)

top_10_publishers_df = top_10_publishers.reset_index()

sb.barplot(x='publisher', y='title', data=top_10_publishers_df)

plt.xticks(rotation=45, ha='right')

plt.xlabel('Publishers')
plt.ylabel('Number of Books')
plt.title('Number of Books Published by Top 10 Publishers')
plt.show()
```
